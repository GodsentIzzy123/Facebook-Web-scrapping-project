```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
import matplotlib.pyplot as plt
import time
import csv
```

```python
#Set up Splinter
browser = Browser("chrome")
#Set up base url

base_url = "https://www.facebook.com/marketplace/charlotte/search?"
```

```python
#Set up search parameters

min_price = 5000
max_price= 30000
days_listed = 60
min_mileage = 50000
max_mileage = 200000
min_year = 2000
max_year = 2022
transmission = "automatic"
make = "Toyota"
model = "Corolla"

#set up full url

url = f"{base_url}minPrice={min_price}&maxPrice={max_price}&daysSinceListed={days_listed}&query=cars&make={make}&exact=false"
```

```python
browser.visit(url)
```

```python
#Scroll down to load more results

scroll_count = 4
scroll_delay = 2

for _ in range (scroll_count):
    browser.execute_script("window.scrollTo(0,document.body.scrollheight);")
    time.sleep(scroll_delay)
```

```python
#Parse the HTML
html = browser.html

#Create a BeautfulSoup object from the scrapped html

market_soup = soup(html, "html.parser")

#Check if HTML was scrapped correctly
```

```python
market_soup
```

```python
browser.quit()
```

```python
titles_div = market_soup.find_all("div", class_ ='x1gslohp xkh6y0r')
titles_list = [title.text.strip() for title in titles_div]

prices_div = market_soup.find_all("span", class_ = "x78zum5")
prices_list = [price.text.strip() for price in prices_div]
```

```python
titles_list

price_final_list  = titles_list[0::4]
title_main_list = titles_list[1::4]
city_final_list = titles_list[2::4]
mileage_final_list = titles_list[3::4]

price_final_list
title_main_list
#title_main_list
#city_final_list
#len(mileage_final_list)
```

```python
combined_data = zip(title_main_list,price_final_list,city_final_list,mileage_final_list)

with open ("final_car_list.csv", 'w',newline ='') as file :
    writer = csv.writer(file)
    writer.writerow(["Car Name","Price","City","Mileage"])
    writer.writerows(combined_data)
```
